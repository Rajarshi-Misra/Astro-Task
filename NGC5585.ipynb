```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from astropy.io import fits
```

```python
NGC5585= fits.open("frame-u-000752-5-0560.fits")
```

```python
NGC5585
```

```python
NGC5585u=fits.open("frame-u-000752-5-0560.fits")
NGC5585i=fits.open("frame-i-000752-5-0560.fits")
NGC5585g=fits.open("frame-g-000752-5-0560.fits")
```

```python
from astropy.visualization import make_lupton_rgb
```

```python
g = NGC5585g[0].data
i = NGC5585i[0].data
u = NGC5585u[0].data
```

```python
rgb_default = make_lupton_rgb(i, g, u, stretch=1.5, Q=10)
plt.imshow(rgb_default, origin='lower')
plt.savefig("rgbNGC5585.jpg")
```
